```python
for state in column_name_mapping:
    orig_df = pd.read_csv(f"./warn_data/{state}.csv")
    column_name_map = column_name_mapping[state]
    base_columns = ["company_name", "location", "notice_date", "layoff_date", "employees_affected"]
    extra_columns = list(column_name_map["misc"].keys())

    new_df = pd.DataFrame()
    for col in base_columns:
        new_col_name = column_name_map[col]
        if new_col_name is not None:
            new_df[col] = orig_df[new_col_name]
    for col in extra_columns:
        new_df[col] = orig_df[column_name_map["misc"][col]]
    new_df = new_df.dropna(subset=["company_name"])
    new_df.to_csv(f"./formatted_warn/{state}.csv")
```
